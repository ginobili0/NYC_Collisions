```python
import pandas as pd
import folium
from IPython.display import HTML
```

```python
def display(m, height=300):
    """Takes a folium instance and embed HTML."""
    m._build_map()
    srcdoc = m.HTML.replace('"', '&quot;')
    embed = HTML('<iframe srcdoc="{0}" '
                 'style="width: 100%; height: {1}px; '
                 'border: none"></iframe>'.format(srcdoc, height))
    return embed
```

```python
accidentData = pd.read_csv('NYPD_Motor_Vehicle_Collisions.csv')
```

```python
accidentData.head()
```

```python
NYC = (40.79, -73.96)

map = folium.Map(location = NYC, zoom_start = 12)

for accident in accidentData[0:100].iterrows():
    map.simple_marker(location = [accident[1]['LONGITUDE'], accident[1]['LATITUDE']], clustered_marker = True)
    
display(map)
```

```python
type(accidentData['DATE'][1])
```

```python
accidentData['DATETIME'] = accidentData['DATE'] + ' ' + accidentData['TIME']

accidentData.drop(['DATE', 'TIME'], axis = 1, inplace = True)

accidentData.head()
```

```python
accidentData['DATETIME'] = pd.to_datetime(accidentData['DATETIME'])
type(accidentData['DATETIME'][1])
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
accidentData['DATETIME'][1].hour
```

```python
accidentData['DATETIME'][1].dayofweek
```

```python
x = []
y = []

for dttime in accidentData['DATETIME']:
    x.append(dttime.hour)
    y.append(dttime.dayofweek)
```

```python
heatmap, x, y = np.histogram2d(x, y)
```

```python
plt.clf()
plt.imshow(heatmap)
plt.show()
```
