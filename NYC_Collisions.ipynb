```python
import pandas as pd
import folium
from IPython.display import HTML
```

```python
def display(m, height=300):
    """Takes a folium instance and embed HTML."""
    m._build_map()
    srcdoc = m.HTML.replace('"', '&quot;')
    embed = HTML('<iframe srcdoc="{0}" '
                 'style="width: 100%; height: {1}px; '
                 'border: none"></iframe>'.format(srcdoc, height))
    return embed
```

```python
NYC = (40.79, -73.96)

map = folium.Map(location = NYC, zoom_start = 12)
```

```python
accidentData = pd.read_csv('NYPD_Motor_Vehicle_Collisions.csv')
```

```python
accidentData.head()
```

```python
for accident in accidentData[0:100].iterrows():
    map.simple_marker(location = accident[1]['LOCATION'], clustered_marker = True)
  
display(map)
```
